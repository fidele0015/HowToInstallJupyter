```python
import pandas as pd
import numpy as np

# Create a sample dataset
products = ['Laptop', 'Mouse', 'Keyboard', 'Monitor', 'Webcam', 'Microphone', 'Headphones', 'Speaker', 'Router', 'Printer']
prices = np.random.randint(50, 1500, size=10)
stock = np.random.randint(10, 200, size=10)

data = {
    'Product': products,
    'Price': prices,
    'Stock': stock
}

df = pd.DataFrame(data)
display(df)
```

```python
import matplotlib.pyplot as plt

# Create a scatter plot
plt.figure(figsize=(10, 6))
plt.scatter(df['Stock'], df['Price'], color='blue', alpha=0.7)
plt.title('Relationship between Stock and Price of Products')
plt.xlabel('Stock Quantity')
plt.ylabel('Price ($)')
plt.grid(True)
plt.show()
```
